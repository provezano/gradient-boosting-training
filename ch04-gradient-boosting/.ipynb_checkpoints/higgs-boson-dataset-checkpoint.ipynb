```python
import numpy as np
import pandas as pd
import xgboost as xgb
```

```python
df = pd.read_csv(r"C:\Users\prove\Downloads\atlas-higgs-challenge-2014-v2.csv.gz",
                nrows=250000, compression='gzip')
```

```python
df.head()
```

```python
del df["Weight"]
del df["KaggleSet"]
df = df.rename(columns={"KaggleWeight": "Weight"})
```

```python
label_col = df['Label']
del df['Label']
df['Label'] = label_col
```

```python
df.head()
```

```python
df.info()
```

```python
df['Label'].replace(('s', 'b'), (1, 0), inplace=True)
```

```python
X = df.iloc[:,1:31]
y = df.iloc[:,-1]
```

```python
df['test_Weight'] = df['Weight'] * 550000 /len(y)
```

```python
s = np.sum(df[df['Label']==1]['test_Weight'])
b = np.sum(df[df['Label']==0]['test_Weight'])
```

```python
xgmat = xgb.DMatrix(X, y, missing=-999.0, weight=df['test_Weight'])
```

```python
param = {}
param['objective'] = 'binary:logitraw'
param['scale_pos_weight'] = b/s
param['eta'] = 0.1
param['max_depth'] = 6
param['eval_metric'] = 'auc'

plst = list(param.items())+[('eval_metric', 'ams@0.15')]
```

```python
watchlist = [(xgmat, 'train') ]
num_round = 120
```

```python
print ('loading data end, start to boost trees')
bst = xgb.train( plst, xgmat, num_round, watchlist )
bst.save_model('higgs.model')
print ('finish training')
```
